```python
avg_price_by_fuel = df.groupby('fuel')['avg_price_brl'].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
avg_price_by_fuel.plot(kind='bar', color=['skyblue', 'lightcoral', 'lightgreen'])
plt.title('Average Car Price by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Average Price (BRL)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
avg_price_by_year_model = df.groupby('year_model')['avg_price_brl'].mean().sort_index()

plt.figure(figsize=(12, 6))
avg_price_by_year_model.plot(kind='line', marker='o', color='dodgerblue')
plt.title('Average Car Price Trend by Model Year')
plt.xlabel('Model Year')
plt.ylabel('Average Price (BRL)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['avg_price_brl'], bins=50, color='teal', edgecolor='black')
plt.title('Distribution of Average Car Prices (BRL)')
plt.xlabel('Average Price (BRL)')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
